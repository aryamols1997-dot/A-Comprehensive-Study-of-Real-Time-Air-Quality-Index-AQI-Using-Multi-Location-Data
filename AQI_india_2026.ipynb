```python
pm25_breakpoints = [
    (0, 30, 0, 50),
    (31, 60, 51, 100),
    (61, 90, 101, 200),
    (91, 120, 201, 300),
    (121, 250, 301, 400),
    (251, 500, 401, 500)
]

pm10_breakpoints = [
    (0, 50, 0, 50),
    (51, 100, 51, 100),
    (101, 250, 101, 200),
    (251, 350, 201, 300),
    (351, 430, 301, 400),
    (431, 600, 401, 500)
]

so2_breakpoints = [
    (0, 50, 0, 50),
    (51, 100, 51, 100),
    (101, 250, 101, 200),
    (251, 350, 201, 300),
    (351, 430, 301, 400),
    (431, 600, 401, 500)
]
no2_breakpoints = [
    (0, 50, 0, 50),
    (51, 100, 51, 100),
    (101, 250, 101, 200),
    (251, 350, 201, 300),
    (351, 430, 301, 400),
    (431, 600, 401, 500)
]

co_breakpoints = [
    (0, 50, 0, 50),
    (51, 100, 51, 100),
    (101, 250, 101, 200),
    (251, 350, 201, 300),
    (351, 430, 301, 400),
    (431, 600, 401, 500)
]  

o3_breakpoints = [
    (0, 50, 0, 50),
    (51, 100, 51, 100),
    (101, 250, 101, 200),
    (251, 350, 201, 300),
    (351, 430, 301, 400),
    (431, 600, 401, 500)
]
nh3_breakpoints = [
   (0, 50, 0, 50),
    (51, 100, 51, 100),
    (101, 200, 101, 200),
    (201, 300, 201, 300),
    (301, 400, 301, 400),
    (401, 500, 401, 500)
]
```

```python
def calculate_subindex(C, breakpoints):
    for B_lo, B_hi, I_lo, I_hi in breakpoints:
        if B_lo <= C <= B_hi:
            return ((I_hi - I_lo) / (B_hi - B_lo)) * (C - B_lo) + I_lo
    return None
```

```python
def get_subindex(row):
    pollutant = row['pollutant_id']
    value = row['pollutant_avg']
    
    if pollutant == 'PM2.5':
        return calculate_subindex(value, pm25_breakpoints)
    elif pollutant == 'PM10':
        return calculate_subindex(value, pm10_breakpoints)
    elif pollutant == 'SO2':
        return calculate_subindex(value, so2_breakpoints)
    elif pollutant == 'NO2':
        return calculate_subindex(value, no2_breakpoints)
    elif pollutant == 'CO':
        return calculate_subindex(value, co_breakpoints)
    elif pollutant == 'OZONE':
        return calculate_subindex(value, o3_breakpoints)
    elif pollutant == 'NH3':
        return calculate_subindex(value, nh3_breakpoints)
    else:
        return None

df['sub_index'] = df.apply(get_subindex, axis=1)
```

```python
aqi_df = df.groupby(['state','city','station'])['sub_index'].max().reset_index()
aqi_df.rename(columns={'sub_index': 'AQI'}, inplace=True)
```

```python
import pandas as pd

bins = [0,50,100,200,300,400,500]
labels = ["Good","Satisfactory","Moderate","Poor","Very Poor","Severe"]

aqi_df['AQI_Category'] = pd.cut(aqi_df['AQI'], bins=bins, labels=labels)
```

```python
import pandas as pd

def categorize_aqi(aqi):
    if pd.isna(aqi):
        return "Unknown"
    elif aqi <= 50:
        return "Good"
    elif aqi <= 100:
        return "Satisfactory"
    elif aqi <= 200:
        return "Moderate"
    elif aqi <= 300:
        return "Poor"
    elif aqi <= 400:
        return "Very Poor"
    else:
        return "Severe"

aqi_df['AQI_Category'] = aqi_df['AQI'].apply(categorize_aqi)
```
